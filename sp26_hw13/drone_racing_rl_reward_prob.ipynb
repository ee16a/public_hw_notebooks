### Visualizations: Parameter Sweeps
This cell convolves the impulse responses you defined above, computes their numerical DTFT, and sweep through different values of $M$ and $N$ to visualize how the window sizes affect the frequency response.

```python
M, N = 4, 3
alpha, beta = 1.0, 1.0
omega = np.linspace(-np.pi, np.pi, 1000)

# 1. Retrieve impulse responses
h_c = h_cmd(M)
h_p = h_prog()
g = g_smooth(N)

# 2. Convolve with smoothing filter
h_cmd_eff = np.convolve(g, h_c)
h_prog_eff = np.convolve(g, h_p)

# 3. Combine to get total impulse response (pad to equal length)
len_tot = max(len(h_cmd_eff), len(h_prog_eff))
h_cmd_eff_padded = np.pad(h_cmd_eff, (0, len_tot - len(h_cmd_eff)))
h_prog_eff_padded = np.pad(h_prog_eff, (0, len_tot - len(h_prog_eff)))
h_total = alpha * h_cmd_eff_padded + beta * h_prog_eff_padded

# 4. Calculate DTFT from the actual convolved arrays
n_tot = np.arange(len(h_total))
H_total_impulse = np.array([np.sum(h_total * np.exp(-1j * w * n_tot)) for w in omega])

# 5. Calculate DTFT from the closed-form mathematical formula
def calc_H_total_formula(M, N, alpha, beta, w_array):
    w = np.where(w_array == 0, 1e-12, w_array)
    H_c = (1 - np.exp(-1j * w * M)) / (1 - np.exp(-1j * w))
    H_p = 1 - np.exp(-1j * w)
    G = (1/N) * (1 - np.exp(-1j * w * N)) / (1 - np.exp(-1j * w))
    return G * (alpha * H_c + beta * H_p)

def m_moving_average_mag(M_val, w_array):
    """Magnitude of standard M-point moving average."""
    w = np.where(w_array == 0, 1e-12, w_array)
    return np.abs((1/M_val) * np.sin(M_val * w / 2) / np.sin(w / 2))

plt.figure(figsize=(15, 5))

# 2. Total frequency response (Constant N=3, Varying M)
plt.subplot(1, 2, 1)
for m_val in [2, 3, 5]:
    mag = calc_H_total_formula(m_val, 3, alpha, beta, omega)
    plt.plot(omega, np.abs(mag), label=f'M={m_val}, N=3')
plt.title('|H_{total}(ω)| (Constant N=3)')
plt.xlabel('ω (rad/sample)')
plt.legend()
plt.grid(True, alpha=0.4)

# 3. Total frequency response (Constant M=3, Varying N)
plt.subplot(1, 2, 2)
for n_val in [2, 3, 5]:
    mag = calc_H_total_formula(3, n_val, alpha, beta, omega)
    plt.plot(omega, np.abs(mag), label=f'M=3, N={n_val}')
plt.title('|H_{total}(ω)| (Constant M=3)')
plt.xlabel('ω (rad/sample)')
plt.legend()
plt.grid(True, alpha=0.4)

plt.tight_layout()
plt.show()
```
