```python
%pylab inline
import numpy as np
import csv
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
def least_squares(A, b):
    return np.linalg.inv(A.T.dot(A)).dot(A.T).dot(b)
```

```python
lines = []
with open('fb.csv', 'r') as f:
    for line in f.readlines():
        l,name = line.strip().split(',')
        lines.append((l,name))

len(lines)

number_of_data_points = len(lines)
fb_stock_data = np.zeros([number_of_data_points,2])
for i in range(number_of_data_points):
    fb_stock_data[i,1] = float(lines[number_of_data_points - 1 - i][1])
    fb_stock_data[i,0] = i + 1
```

```python
plt.plot(fb_stock_data[:,0], fb_stock_data[:,1], 'g^')
```

```python
a1 = (fb_stock_data[:,0]).reshape([number_of_data_points,1])
a2 = np.ones([number_of_data_points,1])
A = np.hstack((a1,a2))
b = (fb_stock_data[:,1]).reshape([number_of_data_points,1])
```

```python
x = least_squares(A,b)
m = x[0]
c = x[1]
```

```python
y_new = m*(fb_stock_data[:,0]) + c
```

```python
plt.plot(fb_stock_data[:,0], fb_stock_data[:,1], 'g^')
plt.plot(fb_stock_data[:,0], y_new, 'r', linewidth=2.0)
plt.show()
```

```python
predict_vec = [i for i in range(62,70)]
predict_vec
```

```python
prediction = m * predict_vec + c
```

```python
plt.plot(fb_stock_data[:,0], fb_stock_data[:,1], 'g^')
plt.plot(fb_stock_data[:,0], y_new, 'r', linewidth=2.0)
plt.plot(predict_vec, prediction, 'bo', linewidth=2.0)
plt.show()
```
