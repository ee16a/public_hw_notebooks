```python
# load up some libraries!
from numpy import array as vector
from matplotlib.pyplot import subplots, grid, show, xlim, ylim
```

```python
# input
origin = vector((0,0))
v1 = vector((6,4))

# plotting details
fig, ax = subplots( figsize=(10,10) )
ax.set_aspect('equal')
grid()
q = ax.quiver( origin[0],origin[1] , v1[0],v1[1] , units='xy' , scale=1 )
xlim(0,8)
ylim(0,6)
show()
```

```python
# input
origin = vector((0,0))
v1 = vector((1,3))
v2 = vector((6,2))


# plotting details
fig, ax = subplots( figsize=(10, 10) )
ax.set_aspect('equal')
grid()
q = ax.quiver( origin[0],origin[1] , v1[0],v1[1] , units='xy' , scale=1 )
q.set_color('red')
r = ax.quiver( v1[0],v1[1] , v2[0],v2[1] , units='xy' , scale=1 )
r.set_color('blue')
x = ax.quiver( origin[0],origin[1] , v1[0]+v2[0],v1[1]+v2[1] , units='xy' ,scale=1)
xlim(0,8)
ylim(0,6)
show()
```
